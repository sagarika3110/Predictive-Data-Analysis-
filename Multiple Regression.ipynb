```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# Import the excel file and convert it into a pandas dataframe to read it
excel_file = 'ProjectData_ISE201.xlsx'
df = pd.read_excel(excel_file)
df
```

```python
# Replace missing values in the dataset with NaN
df = df.replace('\.+', np.nan, regex=True)
```

```python
df.head()
```

```python
# get the number of missing data points per column
missing_values_count = df.isnull().sum()

# look at the # of missing points in the first ten columns
missing_values_count[0:10]
```

```python
df = df.dropna(how='any')
df = df.reset_index(drop=True)
df
```

```python
df=df.fillna(0)
```

```python
df.head()
```

```python
print(df.shape)
```

```python
# Standardize the data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=[0,1])
df[['WAGE','ERSP','ERNO','NEIN','ASSET','AGE','DEP','RACE','SCHOOL']] = scaler.fit_transform(df[['WAGE','ERSP','ERNO','NEIN','ASSET','AGE','DEP','RACE','SCHOOL']])
df.head()   
```

```python
import matplotlib.pyplot as plt
import pandas as pd
x = df[['SCHOOL']]
y = df['HRS']
plt.scatter(x,y, color='r')
plt.xlabel('SCHOOL')
plt.ylabel('HRS')
plt.show()
```

```python
# Correlation Matrix
import seaborn as sns
sns.pairplot(data=df[['HRS','WAGE','ERSP','ERNO','NEIN','ASSET','AGE','DEP','RACE','SCHOOL']])
```

```python
corr=df.corr()
corr
```

```python
# Check Normality of the dataset
# histogram plot
from matplotlib import pyplot
pyplot.hist(df['SCHOOL'])
pyplot.show()
```

```python
# q-q plot
from statsmodels.graphics.gofplots import qqplot
qqplot(df['WAGE'], line='s')
pyplot.show()
```

```python
# Statistical Normality Tests
# normality test-Shapiro-Wilk Test
from scipy.stats import shapiro
stat, p = shapiro(df['SCHOOL'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
    
```

```python
import statsmodels.api as sm
y = df['HRS']
x = df[['WAGE','ERSP','NEIN','DEP','SCHOOL']]
xlog= np.log(x+1)
x = sm.add_constant(xlog)


# Split the data into training set/testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 101)

# Note the difference in argument order
model = sm.OLS(y_train, x_train).fit()
predictions = model.predict(x_test) # make the predictions by the model

# Print out the statistics
model.summary()
```

```python
from sklearn import metrics  
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, predictions))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, predictions))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, predictions))) 
```
